```python
avg_hike = data['PercentSalaryHike'].mean()
```

```python
print('Average salary percentage hike of an employee is:-',avg_hike.round(3),'%')
```

```python
avg_salary_hike_department = data.groupby('Department')['PercentSalaryHike'].mean()
plt.figure(figsize=(6,4))
myexplode1 = [0,0.2,0]
plt.pie(avg_salary_hike_department, labels = avg_salary_hike_department.index, explode = myexplode1,
        colors = ['#0081a7','#98c1d9','#e0fbfc'],autopct='%1.1f%%', shadow = True)
plt.title('Average salary hike by department')
plt.show()
```

- Research & Development department is slight high avg salary hike among other 2 departments
- Sales department is 2nd highest which is 33.4% 

```python
plt.figure(figsize=(8,6))
sns.barplot(x = 'Salary', y = 'JobRole', data = data, color = '#0081a7', orient = 'h', order = None)
plt.show()
```

```python
avg_salary_hike = data.groupby('Gender')['PercentSalaryHike'].mean()
print('Average salary hike by gender:',avg_salary_hike)

fig = plt.figure(figsize=(6,6))


# Plotting the avg salary hike by gender
data1 = data.groupby('Gender')['PercentSalaryHike'].sum()
ax1 = data1.plot(kind = 'bar', color=['#0081a7', '#00afb9'])
ax1.bar_label(ax1.containers[0])
plt.xticks(rotation = 0)
plt.xlabel('Gender')
plt.ylabel('Total Salary Hike')
plt.title('Total Salary Hike by Gender')

# Display the bar chart
plt.show()

```

We can see that Total salary hike for male employees is higher compared to female employees

```python
attrition_rate = data.groupby('Attrition')['Gender'].value_counts(normalize = True).unstack()
attrition_per = attrition_rate*100
ax = attrition_per.plot(kind='bar', color=['#457b9d', '#a8dadc'])
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[0])
plt.xticks(rotation = 0)
plt.xlabel('Gender')
plt.ylabel('Attrition Rate')
plt.title('Attrition Rate by Gender')

plt.show()
```

- Attrition rate is high for male employees than female employees
- Although male employees salary hike is more than female employees
